# Simulate Geometric Brownian Motion

```python
def geometric_brownian(S0, mu, sigma, T: tc.tensor):
    normal  = dist.Normal(loc = tc.tensor(0.0), scale = tc.tensor(1.0))
    S = tc.empty(T)
    S[0] = tc.tensor(S0)
    for i in range(1, T):
        normal_sample = normal.sample()
        S[i] = S[i-1] + S[i-1]*mu*1 + S[i-1]*sigma*1*normal_sample
    return S
```

```python
nsteps = 200
```

```python
gb_stationary = [geometric_brownian(1, 0, 0.2, 20).numpy() for i in range(nsteps)]
gb_upward = [geometric_brownian(1, 0.2, 0.2, 20).numpy() for i in range(nsteps)]
```

## Calculate Vanila Option Prices - BSM

```python
def cal_d1(params):
    sigma, S0, K, r, T = params
    d1 = (tc.log(S0/K) + (r + tc.pow(sigma, 2)/ 2) * T)/(sigma*tc.sqrt(T))
    
    return d1

def cal_d2(params):
    sigma,_,_,_,T = params
    d1 = cal_d1(params)
    d2 = d1 - sigma*tc.sqrt(T)
    return d2
```

```python
def calculate_bsm_option_price(params, option_type = "call", dividend = None, q = 0):
    '''
    `params` is in the order of [sigma, S0, K, r, T]
    '''
    sigma, S0, K, r, T = params
    d1 = cal_d1(params)
    d2 = cal_d2(params)
    def N(x): return(dist.Normal(tc.tensor(0.), tc.tensor(1.)).cdf(x))
    
    price = 0
    
    match option_type:
        
        case "call":
            price = S0*N(d1) - K*tc.exp(-r*T)*N(d2)
        
        case "put":
            price = K*tc.exp(-r*T)*N(-d2) - S0*N(-d1)
            
    return price
```

## Calculate Vanila Option Prices - Binomial Model

```python
class binomial_tree:
    
    def __init__(self, name, path, value, params, depth, children, option_type = "call", eu_am = "European"):
        self.name = name
        self.path = path
        self.value = value
        self.params = params
        self.depth = depth
        self.children = children
        
        self.option_type = option_type
        self.eu_am = eu_am
        
        self.get_tree_params()
        

    def get_tree_params(self):
        self.sigma, self.S0, self.K, self.r, self.q, self.T = self.params
        self.a = tc.exp((self.r - self.q) * self.T)
        self.u = tc.exp(self.sigma * tc.sqrt(self.T))
        self.d = 1/self.u
        self.p = (self.a - self.d)/(self.u - self.d)
        
    
    def build_tree_to_depth(self):
        
        if(self.depth == 0):
            
            multiplyer = tc.prod(tc.tensor(self.path))
            
            St = self.S0 * multiplyer
            
            
            if self.option_type == "call":
            
                if St < self.K:
                    value = 0
                else:
                    value = St - self.K
                    
            elif self.option_type == "put":
                
                if St > self.K:
                    value = 0
                else:
                    value = self.K - St
                    
            self.value = value
                

        elif(self.depth > 0):
            print(self.p)
            
            t1_path = self.path + [self.u]
            t1 = binomial_tree(self.name + "_u", t1_path, self.value, self.params, self.depth-1, [],
                              self.option_type, self.eu_am)
            t1.build_tree_to_depth()
            
            t2_path = self.path + [self.d]
            t2 = binomial_tree(self.name + "_d", t2_path, self.value, self.params, self.depth-1, [],
                              self.option_type, self.eu_am)
            t2.build_tree_to_depth()

            self.children = [t1, t2]
            children_value = (self.p * t1.value + (1-self.p) * t2.value) * tc.exp(-self.r * self.T)
            
            match self.eu_am:
                case "European":
                    self.value = children_value
                
                case "American":
                    multiplyer = tc.prod(tc.tensor(self.path))
                    if self.option_type == "call": current_value = self.S0 * multiplyer - self.K
                    elif self.option_type == "put": current_value = self.K - self.S0 * multiplyer
                    
                    if (current_value > children_value):
                        print("Early exercies at: " + str(self.name))
                    self.value = max(current_value, children_value)
                    
            if self.name == "root_u_d":
                print("@@@@")
                print(self.p)
                print(children_value)
                    
            
        return
            
        
    def __str__(self):
        
        return str("tree") + str(":") + str(self.name)
        
                           
        
```

# Vanilla Option Greeks

## Delta

```python
def delta_euro_call(params): return cal_d1(params)
def delta_euro_put(params): return 1 - cal_d1(params)
```

### Match with example 19.1

```python
print(delta_euro_call(tc.tensor([0.2, 49, 50, 0.05, 0.3846])))
```

## Theta

```python
def N_prime(x): return tc.exp(dist.Normal(tc.tensor(0.), tc.tensor(1.)).log_prob(x))
def N(x): return dist.Normal(tc.tensor(0.), tc.tensor(1.)).cdf(x)
```

```python
def theta_euro_call(params): 
    sigma, S0, K, r, T = params
    d1 = cal_d1(params)
    d2 = cal_d2(params)
    
    theta = (-S0*N_prime(d1)*sigma)/(2*tc.sqrt(T)) - r*K*tc.exp(-r*T)*N(d2)
    return theta

def theta_euro_put(params): 
    sigma, S0, K, r, T = params
    d1 = cal_d1(params)
    d2 = cal_d2(params)
    
    theta = (-S0*N_prime(d1)*sigma)/(2*tc.sqrt(T)) + r*K*tc.exp(-r*T)*N(-d2)
    return theta
```

### Match with example 19.2

```python
print(theta_euro_call(tc.tensor([0.2, 49, 50, 0.05, 0.3846])))
```

## Gamma

```python
def gamma_euro(params): 
    sigma, S0, K, r, T = params
    return N_prime(cal_d1(params))/(S0*sigma*tc.sqrt(T))
```

### Match with example 19.4

```python
print(gamma_euro(tc.tensor([0.2, 49, 50, 0.05, 0.3846])))
```
