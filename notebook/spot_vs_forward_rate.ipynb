```python
import sys
import torch as tc
import matplotlib.pyplot as plt
```

```python
spot_rate = tc.tensor([3, 4, 4.6, 5, 5.3])
```

```python
def forward_from_spot(spot):
    T = tc.arange(1, len(spot) + 1)
    Rf = [(spot[t]* T[t] - spot[t-1]*(T[t]-1))/(T[t] - T[t-1]) for t in tc.arange(1, len(spot))]
    return Rf
```

```python
forward_from_spot(spot_rate)
```

## Relationship between Spot and Forward Rate

### Spot rate grows linearly

```python
spot_rate = tc.arange(1, 10)
plt.plot(spot_rate, label = "Spot")
plt.plot(forward_from_spot(spot_rate), label = "Forward")
plt.legend()
```

### Spot Rate grows exponentially

```python
spot_rate = tc.exp(tc.arange(1, 20)*0.1)
plt.plot(spot_rate, label = "Spot")
plt.plot(forward_from_spot(spot_rate), label = "Forward")
plt.legend()
```

### Spot Rate grows logrithmically

```python
spot_rate = tc.log(tc.arange(1, 20))
plt.plot(spot_rate, label = "Spot")
plt.plot(forward_from_spot(spot_rate), label = "Forward")
plt.legend()
```

### Spot rate decreases linearly

```python
spot_rate = tc.arange(20, 1, -1)
plt.plot(spot_rate, label = "Spot")
plt.plot(forward_from_spot(spot_rate), label = "Forward")
plt.legend()
```

### Spot rate decreases exponentially

```python
spot_rate = tc.exp(tc.arange(20, 1, -1)*0.1)
plt.plot(spot_rate, label = "Spot")
plt.plot(forward_from_spot(spot_rate), label = "Forward")
plt.legend()
```

### Spot rate decreases logrithmically

```python
spot_rate = tc.log(tc.arange(20, 1, -1))
plt.plot(spot_rate, label = "Spot")
plt.plot(forward_from_spot(spot_rate), label = "Forward")
plt.legend()
```
